```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
DataSet = pd.read_csv(io.BytesIO(uploaded['student_scores.csv']))
```

**Part 1: Data Cleaning**

```python
DataSet.head()
DataSet.info()
DataSet.describe()
```

```python
DataSet
```

```python
DataSet['math_score'].fillna(DataSet['math_score'].mean(), inplace=True)
DataSet['english_score'].fillna(DataSet['english_score'].mean(), inplace=True)
DataSet['science_score'].fillna(DataSet['science_score'].mean(), inplace=True)

print("After Filling Missing Values:")
print(DataSet[['math_score', 'english_score', 'science_score']].head())
```

```python
DataSet.drop(columns=['guardian_contact', 'guardian_name'], inplace=True)

print("After Dropping Columns with Too Many Missing Values:")
print(DataSet.columns)
```

```python
DataSet.drop_duplicates(inplace=True)

print("After Dropping Duplicates:")
print("Total rows:", len(DataSet))
```

```python
DataSet.columns = DataSet.columns.str.strip().str.lower().str.replace(' ', '_')

print("After Standardizing Column Names:")
print(DataSet.columns)
```

```python
DataSet.drop(columns=['address'], inplace=True)

print("After Dropping 'address' Column:")
print(DataSet.columns)
```

We dropped the following columns:

- **guardian_name**: This column had 26 out of 30 missing values, making it unsuitable for analysis.
- **guardian_contact**: Similarly, this column was mostly empty, with only 11 non-null values, offering limited usefulness.
- **address**: While complete, this column was irrelevant to our objective (score analysis) and would only add noise due to its lengthy string data.

Removing these columns improves the clarity and efficiency of our dataset

**Part 2: Data Type Conversion**

```python
DataSet['birth_date'] = pd.to_datetime(DataSet['birth_date'])

print("After Converting 'birth_date' to datetime:")
print(DataSet.dtypes)
```

**Part 3: Scaling and Normalization**

```python
minmax = MinMaxScaler()
DataSet[['math_score_norm', 'english_score_norm', 'science_score_norm']] = minmax.fit_transform(
    DataSet[['math_score', 'english_score', 'science_score']]
)

print("After Min-Max Normalization:")
print(DataSet[['math_score_norm', 'english_score_norm', 'science_score_norm']].head())
```

```python
zscore = StandardScaler()
DataSet[['math_score_z', 'english_score_z', 'science_score_z']] = zscore.fit_transform(
    DataSet[['math_score', 'english_score', 'science_score']]
)

print("After Z-score Standardization:")
print(DataSet[['math_score_z', 'english_score_z', 'science_score_z']].head())
```

**Part 4: Encoding Categorical Variables**

```python
le = LabelEncoder()
DataSet['gender_encoded'] = le.fit_transform(DataSet['gender'])

print("After Label Encoding:")
print(DataSet[['gender', 'gender_encoded']].head())
```

We applied encoding techniques to convert categorical values into numerical representations that can be used in data analysis and modeling:

- **Label Encoding** was used for the gender column, which only contains two values ("Male" and "Female"). Since these are binary and non-ordinal, Label Encoding was appropriate and efficient. It is also used when the categories have a natural order (ordinal data) or when there are only two values (binary).

- **One-Hot Encoding** was not applied because the **guardian_name** column which would have been the best candidate was dropped due to excessive missing values (26 out of 30 entries were null). Without another nominal categorical column present, One-Hot Encoding was skipped. It is also used when the categories have no inherent order (nominal data), and you want to avoid assigning arbitrary numeric values.

**Part 5: Feature Engineering**

```python
DataSet['age'] = pd.Timestamp('now').year - DataSet['birth_date'].dt.year

print("After Extracting Age:")
print(DataSet[['birth_date', 'age']].head())
```

```python
DataSet['average_score'] = (DataSet['math_score'] + DataSet['english_score'] + DataSet['science_score']) / 3

print("After Calculating Average Score:")
print(DataSet[['math_score', 'english_score', 'science_score', 'average_score']].head())
```

```python
def categorize(score):
    if score < 70:
        return 'Low'
    elif score <= 85:
        return 'Average'
    else:
        return 'High'

DataSet['score_category'] = DataSet['average_score'].apply(categorize)

print("After Creating Score Category:")
print(DataSet[['average_score', 'score_category']].head())
```

```python
DataSet.to_csv('processed_student_scores.csv', index=False)
print("Processed file saved as 'processed_student_scores.csv'")
```
