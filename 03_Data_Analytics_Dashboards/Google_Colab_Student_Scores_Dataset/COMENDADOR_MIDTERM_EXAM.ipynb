**Part 1: Data Cleaning**

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
ScoresData = pd.read_csv(io.BytesIO(uploaded['student_scores (1).csv']))
```

```python
print("\nStudent Scores Dataset:")
display(ScoresData.head())
```

```python
print("\nDataset Info:")
print(ScoresData.info())
```

```python
print("\nStatistical Description:")
print(ScoresData.describe(include = 'all'))
```

```python
ScoresData['math_score'].fillna(ScoresData['math_score'].mean(), inplace = True)
ScoresData['english_score'].fillna(ScoresData['english_score'].mean(), inplace = True)
ScoresData['science_score'].fillna(ScoresData['science_score'].mean(), inplace = True)
```

```python
duplicate_count = ScoresData.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
ScoresData.drop(columns = ['address','guardian_contact', 'guardian_name'], inplace = True)
display(ScoresData.head())

explanation = """
I dropped the columns: address, guardian_contact, and guardian_name because my data solely focuses on the students grades.
The mentioned columns were not helpful in computing or encoding the students' grades for this data.
"""

print(explanation)
```

```python
ScoresData.columns = ScoresData.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print("\nCleaned Dataset Preview:")
display(ScoresData.head())
```

**Part 2: Data Type Conversion**

```python
ScoresData['birth_date'] = pd.to_datetime(ScoresData['birth_date'], errors = 'coerce')
display(ScoresData.head())
```

```python
object_columns = ScoresData.select_dtypes(include = 'object').columns
potential_numeric = {}
```

```python
for col in object_columns:
    try:
        ScoresData[col] = pd.to_numeric(ScoresData[col])
        potential_numeric[col] = 'converted to numeric'
    except ValueError:
        potential_numeric[col] = 'not converted (text data)'
```

```python
print("\nConversion Results:")
for col, status in potential_numeric.items():
    print(f"{col}: {status}")

print("\nUpdated Data Types:")
display(ScoresData.dtypes)
```

**Part 3: Scaling and Normalization**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
score_columns = ['math_score', 'english_score', 'science_score']
```

```python
minmax_scaler = MinMaxScaler()
ScoresData_minmax = ScoresData.copy()
ScoresData_minmax[score_columns] = minmax_scaler.fit_transform(ScoresData[score_columns])

display(ScoresData_minmax[score_columns].head())
```

```python
zscore_scaler = StandardScaler()
ScoresData_zscore = ScoresData.copy()
ScoresData_zscore[score_columns] = zscore_scaler.fit_transform(ScoresData[score_columns])

display(ScoresData_zscore[score_columns].head())
```

```python
print("Min-Max Normalized Scores:")
print(ScoresData_minmax[score_columns].head())

print("\nZ-score Standardized Scores:")
print(ScoresData_zscore[score_columns].head())
```

**Part 4: Encoding Categorical Variables**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ScoresData_encoded = ScoresData.copy()
```

```python
label_encoder = LabelEncoder()
ScoresData_encoded['gender_encoded'] = label_encoder.fit_transform(ScoresData_encoded['gender'])
```

```python
print("Encoded DataFrame Preview:")
print(ScoresData_encoded.head())
```

I used **Label Encoding** for the `gender` column since it only has two categories (Male and Female). This method is efficient for binary or ordinal data.

I **did not use One-Hot Encoding** because the other categorical columns ( `guardian_name`,`address`,`guardian_contact`) were dropped due to high missing values or uniqueness, which would make one-hot encoding impractical and add unnecessary complexity.

**Part 5: Feature Engineering**

```python
current_year = pd.Timestamp.now().year
ScoresData['birth_year'] = pd.DatetimeIndex(ScoresData['birth_date']).year
ScoresData['age'] = current_year - ScoresData['birth_year']

display(ScoresData[['birth_date', 'age']].head())
```

```python
ScoresData['average_score'] = ScoresData[['math_score', 'english_score', 'science_score']].mean(axis = 1)
```

```python
def score_category(score):
    if score < 70:
        return 'Low'
    elif score <= 85:
        return 'Average'
    else:
        return 'High'

ScoresData['score_category'] = ScoresData['average_score'].apply(score_category)
```

```python
print("Feature Engineered Data Preview:")
display(ScoresData[['name', 'age', 'average_score', 'score_category']].head())
```

**FINAL STUDENT SCORE DATASET**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
z_scores = scaler.fit_transform(ScoresData[['math_score', 'english_score', 'science_score']])

ScoresData['math_zscore'] = z_scores[:, 0]
ScoresData['english_zscore'] = z_scores[:, 1]
ScoresData['science_zscore'] = z_scores[:, 2]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(ScoresData[['math_score', 'english_score', 'science_score']])

ScoresData['math_minmax'] = minmax_scaled[:, 0]
ScoresData['english_minmax'] = minmax_scaled[:, 1]
ScoresData['science_minmax'] = minmax_scaled[:, 2]
```

```python
display(ScoresData)
```
